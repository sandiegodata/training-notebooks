### Required Tools

The first thing you will have to do is install the basic tools: 

**Anaconda python distribution from Continuum Analytics.**

The Anaconda distribution provides the most common Python packages for scientific computing, and ensures that we have a consistent environment. You can configure it to work along side of other Python distributions, so you only need to use it when you are working on Data Library projects. Anaconda comes with its own version of Python 2.7, IPython Notebook, plus other libraries like matplotlib and a package manager ("conda") that you can use to install other packages or other Python versions.

   * Install Anaconda: http://continuum.io/downloads


**A Github Account**: Git is a distributed source code repository, and Github is a commercial service that uses git to make it really easy to share code. You will need to create an account on Github.com, to use the service. 

  * Get a Github account: http://github.com/

**Install a Git Client**. You can use the command line tools for everything, but it is much easier to use a GUI tool. We recommend Atlassian SourceTree, since it is easy to use, cross-platform, and free, but you can use any alternative you want to. 

   * Get a Github client: https://www.atlassian.com/software/sourcetree/overview
